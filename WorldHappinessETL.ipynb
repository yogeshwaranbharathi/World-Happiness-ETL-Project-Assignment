```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

**`Extract: import world happiness csv files`**

```python
file = "2015.csv"
df2015 = pd.read_csv(file)
df2015.head()
```

```python
file = "2020.csv"
df2020 = pd.read_csv(file)
df2020.head()
```

**Transform: drop columns, rename columns, round decimals**

```python
df2015.columns
```

```python
df2020.columns
```

```python
# drop unwanted columns from 2015 df

drop2015 =df2015.drop(columns =['Region','Standard Error', 'Economy (GDP per Capita)', 'Family','Freedom', 'Dystopia Residual'])
drop2015.head()
```

```python
#drop unwanted columns from 2020 df

drop2020 = df2020.drop(columns = ['Regional indicator', 
       'Standard error of ladder score', 'upperwhisker', 'lowerwhisker',
       'Logged GDP per capita', 'Social support', 
       'Freedom to make life choices',
       'Ladder score in Dystopia',
       'Explained by: Log GDP per capita', 'Explained by: Social support',
       'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption',
       'Dystopia + residual'])
drop2020.head()
```

```python
# rename columns in 2015 dropped df

drop2015_transformed = drop2015.rename(columns={"Country": "country", 
                                                      "Happiness Rank": "happiness_rank_2015", 
                                                      "Happiness Score": "happiness_score_2015",
                                                          "Health (Life Expectancy)": "life_expectancy_2015",
                                                          "Generosity": "generosity_2015", "Trust (Government Corruption)": "trust_2015"})


drop2015_order = drop2015_transformed[["country", "happiness_rank_2015", "happiness_score_2015", "life_expectancy_2015", "generosity_2015", "trust_2015"]]
drop2015_order.head()
```

```python
# rename columns in 2020 dropped df

drop2020_transformed = drop2020.rename(columns={"Country name": "country", 
                                                      "Ladder score": "happiness_score_2020",
                                                        "Healthy life expectancy": "life_expectancy_2020",
                                                          "Generosity": "generosity_2020",
                                                          "Perceptions of corruption": "trust_2020"})
drop2020_transformed.head()
```

Rounded Numerical values to (2) decimal places

```python
# round all numerical columns to two decimal points to clean up output 

def rounding(x):
    return round(x,2)
```

```python
# Rounded Numerical values to (2) decimal places 2015


drop2015_round = drop2015_order
drop2015_round[["happiness_score_2015", "life_expectancy_2015", "generosity_2015", "trust_2015"]] = drop2015_round[["happiness_score_2015", "life_expectancy_2015", "generosity_2015", "trust_2015"]].apply(rounding)
drop2015_round.head()
```

```python
#Rounded Numerical values to (2) decimal places 2020

drop2020_round = drop2020_transformed
drop2020_round[["happiness_score_2020", "life_expectancy_2020", "generosity_2020", "trust_2020"]] = drop2020_round[["happiness_score_2020", "life_expectancy_2020", "generosity_2020", "trust_2020"]].apply(rounding)
drop2020_round.head()
```

**Merged data frames based on country**

```python
merge_df = pd.merge(drop2015_ordered, drop2020_transformed, on="country", how="outer")
merge_df.head(158)
```
